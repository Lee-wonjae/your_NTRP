```python
import cv2
import numpy as np
import os

#      
ref_image_path = 'novak_med_frame/frame_000121.jpg'
ref_image = cv2.imread(ref_image_path)
ref_hist = cv2.calcHist([ref_image], [0, 1, 2], None, [256, 256, 256], [0, 256, 0, 256, 0, 256])

#   
frames_directory = 'novak_med_frame'
os.makedirs('novak_med_over_30', exist_ok=True)
os.makedirs('novak_med_under_30', exist_ok=True)

#      
for frame_file in os.listdir(frames_directory):
    #   
    frame_path = os.path.join(frames_directory, frame_file)
    frame = cv2.imread(frame_path)

    #   
    frame_hist = cv2.calcHist([frame], [0, 1, 2], None, [256, 256, 256], [0, 256, 0, 256, 0, 256])

    #  
    similarity = cv2.compareHist(ref_hist, frame_hist, cv2.HISTCMP_CORREL)

    #   
    if similarity > 0.01:  # ,   
        #   
        print(f"Similar frame found: {frame_file} (Similarity: {similarity})")
        save_path=os.path.join('novak_med_over_30',frame_file)
        cv2.imwrite(save_path, frame)
    
    else :
        save_path=os.path.join('novak_med_under_30',frame_file)
        cv2.imwrite(save_path, frame)
        #           
```
